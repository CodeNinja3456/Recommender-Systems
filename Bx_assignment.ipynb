```python
import pandas
books=pandas.read_csv('BX-Books.csv')
books.head(5)
```

```python
book_details=pandas.read_csv('BX-Book-Ratings.csv')
book_details.head(5)
```

```python
data=pandas.merge(books,book_details,on='ISBN')
data.head(5)
```

```python

mean_rating =pandas.DataFrame(data.groupby('Book-Title')['Book-Rating'].mean()) 

mean_rating['Mean Rating']=pandas.DataFrame(data.groupby('Book-Title')['Book-Rating'].mean()) 
mean_rating.head(10)


mean_rating['Number of Rating']=pandas.DataFrame(data.groupby('Book-Title')['Book-Rating'].count()) 
mean_rating.head(10)
```
